```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'school_code':['s001','s002','s003','s001','s002','s004'],
    'class':['V','V','VI','VI','V','VI'],
    'name':['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'data_of_Birth':['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age':[12,12,13,13,14,15],
    'height':[173,192,168,174,178,185],
    'weight':[35,32,33,30,31,32],
    'address':['street1','street2','street3','street1','street2','street4']},
    index=[0,1,2,3,4,5])
df
```

```python
obj=df.groupby(['age'])
for name,group in obj:
    print(name)
    print(group)
```

```python
obj.agg({'height':['mean','sum','max','min']})
```
